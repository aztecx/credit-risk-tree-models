```python
import pandas as pd

df = pd.read_csv(
    "../data/german.data-numeric",
    sep='\s+',  
    header=None             
)

print("Shape of df:", df.shape)
df.head()
```

```python
# 24 feature columns + 1 target column (last one)
num_features = 24

feature_names = [f"feat_{i}" for i in range(1, num_features + 1)]
all_columns = feature_names + ["target"]

df.columns = all_columns

print(df.columns)
df.head()
```

```python
df["target"].value_counts()
```

```python
print("Info:")
df.info()

print("\nMissing values per column:")
print(df.isna().sum())
```

```python
# Map original target: 1 = good, 2 = bad
# Creating the BadCredit: 1 = bad, 0 = good

df["BadCredit"] = df["target"].map({1: 0, 2: 1})

print("Value counts of BadCredit:")
print(df["BadCredit"].value_counts())
print("\nProportions of BadCredit:")
print(df["BadCredit"].value_counts(normalize=True))
```

```python
df.to_csv("../data/german_credit_numeric_clean.csv", index=False)

print("Saved cleaned dataset to ../data/german_credit_numeric_clean.csv")
```
